```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

## 先获取所有栏目

```python
index = 'http://soushu.dzsxzz.com/'
req = requests.get(index)
req.encoding = 'gbk'
req.encoding = 'utf' # 需要进行两次重编码
soup = BeautifulSoup(req.text) # 获取所有栏目
title = soup.select('.navbar li')
```

```python
columns = {}
for i in title:
    href = i.select('a')[0].attrs.get('href', None)
    columns.setdefault(i.text, href)
```

## 栏目+编号暴力获取

```python
def get_book_name(url):
    req = requests.get(url)
    req.encoding = 'gbk'
    req.encoding = 'utf' # 需要进行两次重编码
    soup = BeautifulSoup(req.text)
    title_li = soup.select('.article-title')
    if len(title_li) == 0:
#         print(url, req.text) # 测试为什么获取不到
        return None
    else:
        return title_li[0].text
```

```python
url = 'http://soushu.dzsxzz.com/{}/{}.html'
columns_li = [] # 栏目名称
columns_id_li = [] # 栏目编号
book_name_li = []
for col in columns.keys():
    if col == '首页':
        continue
    print('正在获取栏目：', col)
    for i in range(1, 2000):  # 一般就到了2000个
#     for i in range(1600, 1700):  # 测试 http://soushu.dzsxzz.com/cyzx/1616.html
        if i % 100 == 0:
            wait_seconds = 3
            print('已经获取100个，休息', wait_seconds, '秒')
            time.sleep(3)
        new_url = url.format(columns[col], i)
        book_name = get_book_name(new_url)
        if book_name == None:
            continue
        else:
            columns_li.append(col)
            columns_id_li.append(i)
            book_name_li.append(book_name)
            print(book_name, i)
```

```python
data = pd.DataFrame({'栏目名称': columns_li, '栏目编号': columns_id_li, '书名': book_name_li})
```

```python
data.to_excel('搜书网2021年11月22日.xlsx')
```
